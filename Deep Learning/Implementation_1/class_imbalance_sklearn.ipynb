```python
import pandas as pd
import numpy as np
import os 
import glob
from sklearn.utils import resample
```

```python
ROOT = '/home/trojan/skia_projects/3d_facial_segmentation/part_segmentation/pointnet/pointnet/datasets/head_data'
```

```python
for idx, item in enumerate(glob.glob(ROOT + '/train/*.txt')):
    df = pd.read_csv(item, delimiter=' ', index_col=False, names = ['x', 'y', 'z', 'label'])
    #print(df)
    df_majority = df[df.label == 0]
    df_minority = df[df.label == 1]
    #print(len(df_majority), len(df_minority))
    df_minority_upsampled = resample(df_minority,
                                    replace=True,
                                    n_samples=len(df_majority),
                                    random_state=123)
    #print(len(df_majority), len(df_minority_upsampled))
    df_balanced = pd.concat([df_majority, df_minority_upsampled])
    #print(df_balanced)
    #print(df_balanced.label.value_counts())
    df_balanced.to_csv(os.path.join(ROOT, 'train_balanced', f'{idx+1}.txt'), sep=' ', mode='w', index=False, header=None)
```
